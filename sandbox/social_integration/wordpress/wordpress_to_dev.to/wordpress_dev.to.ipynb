```python
import requests
import markdownify as md
from config import Wordpress
from config import DevTo
```

```python
WORDPRESS_URL = 'https://public-api.wordpress.com/'
AUTH_TOKEN_URL = f'{WORDPRESS_URL}oauth2/token'
POSTS_URL = f'{WORDPRESS_URL}rest/v1.1/sites/{Wordpress.SITE_ID}/posts/{Wordpress.POST_ID}'
DEVTO_API_KEY = DevTo.API_KEY
```

```python
auth_message_body = {
    'client_id': Wordpress.CLIENT_ID,
    'client_secret': Wordpress.CLIENT_SECRET,
    'grant_type': Wordpress.GRANT_TYPE,
    'username': Wordpress.USERNAME,
    'password': Wordpress.APP_PASSWORD
}

response = requests.post(AUTH_TOKEN_URL, data=auth_message_body)
response_json = response.json()

ACCESS_TOKEN = response_json['access_token']
```

```python
# Get Post content
headers = {
    'Authorization': f'Bearer {ACCESS_TOKEN}'
}

posts_response = requests.get(POSTS_URL, headers=headers)
posts_response_json = posts_response.json()
```

```python
with open("wordpress_rersponse.html", "w") as outfile:
    outfile.write(str(posts_response_json['content']))
```

```python
content_md = md.markdownify(posts_response_json['content'], heading_style="ATX")
```

```python
url = 'https://dev.to/api/articles'

headers = {
    'api-key' : f'{DEVTO_API_KEY}',
    'content' : 'application/json'
}

payload = {
  'article': {
    'title': f"{posts_response_json['title']}",
    'published': True,
    'body_markdown': f"{content_md}",
    'tags': [
      'python',
      'rest',
      'api'
    ],
    'series': 'Social Networking APIs',
    'canonical_url' : f"{posts_response_json['URL']}",
    'social_image' : f"{posts_response_json['featured_image']}",
    'reading_time_minutes' : 7
  }
}
```

```python
r = requests.post(url, headers=headers, json=payload)

r.status_code
r.json()
```
